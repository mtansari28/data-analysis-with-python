## 7: Recapping What I've Learned

Task:
To filter an array, take a random sample from that new array, then use .where to categorise the data into a final array (by mean and median).

```python
import pandas as pd
import numpy as np

retail_df = pd.read_csv(
    "path_to_dataset", skiprows=range(1, 11000), nrows=1000
)

family_array = np.array(retail_df["family"])
sales_array = np.array(retail_df["sales"])
```

```python
mask = (family_array == "PRODUCE")

produce_array = sales_array[mask]
```

```python
rng = np.random.default_rng(2022)

random_array = rng.random(30)

sampled_array = produce_array[random_array < 0.5]
```

```python
mean = sampled_array.mean()

median = np.median(sampled_array)

print(f"Mean: {sampled_array.mean()}")

print(f"Median: {np.median(sampled_array)}")
```

```python
#My mistake here was I wrote a > instead of a >=, leaving the actual median case uncategorised.

final_array = np.where((sampled_array > mean) & (sampled_array > median),
                       "above_both",
                       np.where((sampled_array >= median), "above_median", "below_both"))

final_array
```
